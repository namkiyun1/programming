```python
year2017=df.iloc[:,[0,2,3]]
year2017=year2017.reset_index()
year2017
```

```python
year2017
year2017 = year2017[year2017['행정구역별(시군구)']!="전국"]
year2017 = year2017[year2017['행정구역별(시군구)']!="읍부"]
year2017 = year2017[year2017['행정구역별(시군구)']!="면부"]
year2017 = year2017[year2017['행정구역별(시군구)']!="동부"]
```

각 지역들 마다 1인가구가 어느 곳이 제일 많고, 적고를 보여준다.

```python
(
year2017[['행정구역별(시군구)','man_2017','girl_2017']]
          .sort_values(['man_2017','girl_2017'], ascending=[False,False])
          .drop_duplicates(subset=['행정구역별(시군구)'])     
)
```

```python
year_2017=year2017.set_index('행정구역별(시군구)')
```

```python
#1인가구 분포율 궁금한 지역 입력
name = input('1인가구 분포율 어느 지역이 궁금하십니까?ex)경기도')#지역 이름 입력받는다.
a=year_2017.index.str.contains(name)#행을 찾아 해당 지역 인구를 a라는 변수에 저장
enter=year_2017[a] #a라는 값을 enter 에 다시 저장합니다.
#enter=enter[enter["연령별"]!="합계"] #연령별 합계를 다 삭제합니다.

enter #name 지역의 값을 표시합니다.
```

```python
mangirl=enter.iloc[:,[0,1,2]]# [:,[0,2,3]]는 중간의 콤마에서 앞에는 행 뒤는 열입니다 따라서, 0,2,3는 연령별 , 여자,남자만 출력합니다.
mangirl=mangirl.set_index('age') # 표를 그래프로 나타내야 하기 때문에 바꿉니다. 인덱스를 연령별로

#del mangirl['행정구역별(시군구)'] #열의 행정구역별 시군구를 삭제합니다
mangirl.T #행과 열을 바꿉니다
```

```python
import matplotlib.pyplot as plt #표 출력하기 위해 임포트 해줍니다
plt.rc('font', family="Malgun Gothic") #폰트 설정
plt.rcParams['font.size']=30 #폰트크기 설정
 
plt.figure(figsize=(30,15))#그래프 사이즈 x축 30, y축 15로 설정

plt.plot(mangirl['man_2017'],linewidth=3,label="남자")# 꺾은선 그래프로 출력합니다
plt.plot(mangirl['girl_2017'],linewidth=3,label="여자")# 꺾은선 그래프로 출력합니다

plt.xlabel('연령대') #아래 x축 제목 표시
plt.ylabel('인구 수')#아래 y축 제목 표시

plt.title(name+'지역의 1인가구 현황') #name의 지역의 1인가구 현황을 보여줍니다. 표 상단의 제목
plt.legend()
plt.show()#코딩한 그래프를 그리게 합니다.
```
